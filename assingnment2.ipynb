```python
# Exercise 1
# Create a dictionary for one student with name, age, and a nested dictionary of three subjects and marks
# Print the student’s name and age, then calculate and print the average mark
student = {
    "student_id_001": {
        "name": "Alice",
        "age": 20,
        "marks": {
            "Math": 85,
            "Science": 90,
            "History": 78   
        }
    },
    "student_id_002": {
        "name": "Bob",
        "age": 22,
        "marks": {
            "Math": 88,
            "Science": 76,
            "History": 92   
        }
    },
    "student_id_003": {
        "name": "Charlie",
        "age": 19, 
        "marks": {
            "Math": 95,
            "Science": 89,
            "History": 84   
        }
    }
}
# Print the dictionary
for id, info in student.items():
    # print(f"ID: {id}")
    # print(f"Name: {info['name']}")
    # print(f"Age: {info['age']}")
    # print("Marks:")
    total_marks = 0
    for subject, mark in info['marks'].items():
        # print(f"  {subject}: {mark}")
        total_marks += mark
    student[id]['total_marks'] = total_marks
    average_marks = total_marks // len(info['marks'])
    student[id]['average_marks'] = average_marks
    print(f" Name: {info['name']} Average Marks: {average_marks:.2f}")
    print()
print(student)
```

```python
# Exercise 2:
# # Create a dictionary for one department with keys "budget" and "employees"
# "employees" should be a list of two dictionaries, each with "name" and "salary"
# # Calculate and print the total salary, then print if it is within the budget

Employees = {
     "Sales":{
    "Budget": 500000,
    "Employees":[
        {"Name": "Alice",  "Salary": 80000},
        {"Name": "Bob",    "Salary": 75000},
        {"Name": "Charlie","Salary": 70000}
    ]
    },
    "Engineering":{
    "Budget": 1200000,
    "Employees":[
        {"Name": "David",  "Salary": 95000},
        {"Name": "Eve",    "Salary": 105000},
        {"Name": "Frank",  "Salary": 110000}
    ]
    },
    "HR":{
    "Budget": 30000,
    "Employees":[   
        {"Name": "Grace",  "Salary": 60000},
        {"Name": "Heidi",  "Salary": 65000}
    ]
    }

}

for dep,info in Employees.items():
    print(f"Department: {dep}")
    print(f" Budget: {info['Budget']}")
    total_salary = 0
    for emp in info["Employees"]:
        total_salary += emp['Salary']
    print(f" Total Salary Expense: {total_salary}")
    info['Salary'] = total_salary
    if total_salary > info['Budget']:
        print(" Warning: Salary expenses exceed budget!")
    print()
```

```python
#Exercise 4:

#Create two sets: tech_subscribers and news_subscribers with some overlapping emails
# # Print the emails that are in both sets
# Print all unique emails from both sets
# Remove one email from both sets and print the updated sets


tech_subscribers = {"deepak123@gmail.com","sathu@gmail.com","deepak123@gmail.com","virat@gmail.com"}
new_subscribers = {"devillers@mail.com","devillers@mail.com","brevis@mail.com","deepak123@gmail.com"}


#Print all unique email addresses from both sets using a loop
all_subscribers = tech_subscribers.union(new_subscribers)
for subscriber in all_subscribers:
    print(subscriber)

# #Print email addresses that are in both sets
common_subscribers = tech_subscribers.intersection(new_subscribers)
print(common_subscribers)

# Remove a specific subscriber safely
all_subscribers.discard("devillers@mail.com")

print("After removing:", all_subscribers)
```
