```python
# Exercise 1
# Write a function `count_long_lines(filename, limit)` that:
# 1. Opens a text file and reads all lines.
# 2. Uses a **for-loop** to count how many lines have length > limit.
# 3. Uses an **if-statement** inside the loop to check line length.
# 4. Returns the count.
# Then, write code that:
# - Prompts the user for a filename and a limit.
# - Calls the function.
# - Prints the result.
def count_log_lines(file_name,limit):
    counter=0
    with open('dest.txt','r') as file:
        data=file.readlines()
        for line in data:
            word_length = line.split()
            # print(f"Word Length : {len(word_length)}")
            line_length=len(line)
            if line_length > limit:
                counter+=1
        return counter

limit=7
file_name='dest.txt'
print(count_log_lines(file_name,limit))
```

```python
# Exercise 2:
# Write a script that:
# 1. Opens a file named "numbers.txt" in write mode.
# 2. Uses a **while-loop** to repeatedly ask the user for integers.
# 3. Stops when the user enters "stop".
# 4. Writes each valid integer to the file (one per line).
# After writing:
# - Re-open the file in read mode.
# - Use a for-loop to read each number and compute the sum.
# - Print the total sum.
def count_sum(text_file):
    total = 0
    with open(text_file, 'r') as file:
        for line in file:
            line = line.strip()
            if line.isdigit():        # safety check
                total += int(line)
    return total


with open('dc.txt', 'a') as file:
    while True:
        user_input = input("Enter a number (enter 5 to stop): ").strip()
        if user_input == '5':
            print("Sum:", count_sum('dc.txt'))
            break
        file.write(user_input + '\n')




    
            


```

```python
# Exercise 3:
# Create a function `filter_words(input_file, output_file, min_length)` that:
# 1. Opens input_file, reads all lines.
# 2. For each line, splits into words.
# 3. Uses a loop + if-statement to keep only words whose length >= min_length.
# 4. Writes these filtered words into output_file (one line per original line).
# Then call the function using any sample input/output filenames.

def filter_words(input_file,output_file,min_length):
    with open(input_file,'r') as file:
        for line in file:
            line=line.strip()
            words=line.split(' ')
            for i in words:
                if len(i)>min_length:
                    write_output_file(output_file,i)

def write_output_file(output_file,word):
    with open(output_file,'a') as file:
        file.write(word+ '\n')
        

input_file='dest.txt'
output_file='example.txt'
min_length=2
filter_words(input_file,output_file,min_length)
```

```python
# Exercise 4:
# Write code that:
# 1. Opens a file "temperatures.txt" which contains one temperature per line.
# 2. Reads all lines using readlines().
# 3. Converts them to floats.
# 4. Uses a function `average_temp(data)` to compute the average.
# 5. Prints:
# - The average temperature
# - "High" if avg > 25
# - "Moderate" if 15 <= avg <= 25
# - "Low" if avg < 15

def avg_temp(x):
    if x>25:
        print('High')
    elif x<=25 and x>=15:
        print('Moderate')
    elif x<15:
        print('Low')

with open("temp.txt",'r') as file:
    content=file.readlines()
    for i in content:
        x=float(i.replace('C',''))
        avg_temp(x)

        

```

```python
# Exercise 5:
# Write a program that:
# 1. Opens a file "logs.txt" and reads each line.
# 2. Each line is formatted like: "LEVEL: message"
# 3. Use a function `count_level(filename, level)` that:
# - Opens the file
# - Counts how many lines start with the given level (e.g., "ERROR")
# - Uses loops + if to check matching lines
# 4. Print how many ERROR, WARNING, and INFO lines are in the file.

def clean_string(line):
    if ":" not in line:
        return None      # invalid line
    return line.split(":", 1)[1].strip()

log_count = {"ERROR": 0, "WARNING": 0, "INFO": 0}

def counter(level):
    if level in log_count:
        log_count[level] += 1

with open("logs.txt", "r") as file:
    for line in file:        # memory efficient
        level = clean_string(line)
        counter(level)
        
print("ERROR count:", log_count["ERROR"])



```
